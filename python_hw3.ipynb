# Задание 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("train.csv")
```

# Задание 2

```python
for col in ["Survived", "Pclass", "Sex", "Parch"]:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(data=df, x="Age", bins=20)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# Заданеи 3

```python
plt.figure(figsize=(6,4))

sns.boxplot(x=df["Age"])

plt.title("Boxplot of Passenger Age")
plt.xlabel("Age")

plt.show()
```

# Задание 4

На графике видно, что большинство пассажиров имеют возраст примерно от 20 до 40 лет.
Медианный возраст находится около 28–30 лет.
Также наблюдаются выбросы среди пассажиров 60–80 лет и очень маленькие значения возраста, соответствующие младенцам.

# Задание 5

```python
survived_counts = df["Survived"].value_counts()
pclass_counts = df["Pclass"].value_counts()

plt.figure(figsize=(6, 6))
survived_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=survived_counts.index
)
plt.title("Survival Rate of Passengers")
plt.ylabel("")
plt.show()

plt.figure(figsize=(6, 6))
pclass_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=pclass_counts.index
)
plt.title("Passenger Class Distribution")
plt.ylabel("")
plt.show()
```

# Задание 6

```python
numeric_df = df.select_dtypes(include=["int64","float64"])

sns.pairplot(numeric_df)

plt.suptitle("Pairplot of Numeric Variables", y=1.02)

plt.show()
```

# Задание 7

```python
plt.figure(figsize=(6,5))

sns.violinplot(data=df, x="Sex", y="Age")

plt.title("Age Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Age")

plt.show()
```

# Задание 8

```python
plt.figure(figsize=(8,6))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix of Numerical Features")

plt.show()
```

# Задание 9

```python
fig = px.sunburst(
    df,
    path=["Pclass", "Sex"],
    title="Passenger Distribution by Class and Sex",
)

fig.update_traces(textinfo="label+percent entry+value")

fig.show()
```
