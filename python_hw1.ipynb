**Задание 1**

```python
def parse_int(s: str) -> int | None:
    try:
        return int(s.strip())
    except ValueError:
        return None

test_data = ["12", " -7 ", "12a", ""]

for item in test_data:
    res = parse_int(item)
    print(f"Ввод: '{item}' -> Результат: {res}")
```

**Задание 2**

```python
import random

def count_positive(nums: list[int]) -> int:
    count = 0
    for n in nums:
        if n > 0:
            count += 1
    return count

def sum_even(nums: list[int]) -> int:
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total

def first_divisible(nums: list[int], k: int) -> int | None:
    for n in nums:
        if n % k == 0:
            return n
    return None

numbers = [random.randint(-50, 50) for _ in range(20)]

print(f"Список чисел: {numbers}")
print(f"Количество положительных: {count_positive(numbers)}")
print(f"Сумма чётных: {sum_even(numbers)}")

k = 5
print(f"Первое число, кратное {k}: {first_divisible(numbers, k)}")
```

**Задание 3**

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]

sorted_by_price = sorted(items, key=lambda x: x["price"])
sorted_by_rating = sorted(items, key=lambda x: x["rating"], reverse=True)

print("Сортировка по цене (возрастание):")
for item in sorted_by_price:
    print(item)

print("\nСортировка по рейтингу (убывание):")
for item in sorted_by_rating:
    print(item)
```

**Задание 4**

```python
class Chocolate:
    def __init__(self, name: str, price: float):
        if price <= 0:
            raise ValueError("Цена должна быть больше нуля")
        self.name = name
        self.price = price

    def discount(self, percent: int) -> float:
        return self.price * (1 - percent / 100)

try:
    choc1 = Chocolate("Alpen Gold", 100.0)
    choc2 = Chocolate("Ritter Sport", 250.0)

    print(f"Шоколад: {choc1.name}, Цена без скидки: {choc1.price}")
    print(f"Цена со скидкой 20%: {choc1.discount(20)}")

    print(f"\nШоколад: {choc2.name}, Цена без скидки: {choc2.price}")
    print(f"Цена со скидкой 15%: {choc2.discount(15)}")

except ValueError as e:
    print(f"Ошибка: {e}")
```
