# Вспомогательные функции, объявление библиотек и переменных

```python
import pandas as pd

def extract_first_name(name):
    if "(" in name:
        start = name.find("(") + 1
        end = name.find(")")
        return name[start:end].split()[0]

    part = name.split(".")[1]
    return part.strip().split()[0]


def get_most_common_name(df, sex):
    counts = df[df["Sex"] == sex]["FirstName"].value_counts()
    return counts.index[0], counts.iloc[0]


def get_most_common_name_in_class(df, pclass, sex):
    counts = df[(df["Pclass"] == pclass) & (df["Sex"] == sex)]["FirstName"].value_counts()
    return counts.index[0], counts.iloc[0]


def count_n_person_cabins(df):
    cabins = df["Cabin"].dropna()
    cabin_counts = cabins.value_counts()
    return cabin_counts.value_counts().sort_index()

df = pd.read_csv("train.csv")
df["FirstName"] = df["Name"].apply(extract_first_name)
```

# Задание 3

```python
print("\nПроцент выживаемости по классам:")

survival_rate = df.groupby("Pclass")["Survived"].mean() * 100
print(survival_rate)
```

# Задание 4

```python
print("\nСамые популярные имена:")

male_name, male_count = get_most_common_name(df, "male")
female_name, female_count = get_most_common_name(df, "female")

print(f"Мужское имя: {male_name} ({male_count} пассажиров)")
print(f"Женское имя: {female_name} ({female_count} пассажиров)")
```

# Задание 5

```python
print("\nПопулярные имена по классам:")

for pclass in sorted(df["Pclass"].unique()):

    male_name, male_count = get_most_common_name_in_class(df, pclass, "male")
    female_name, female_count = get_most_common_name_in_class(df, pclass, "female")

    print(f"\nКласс {pclass}:")
    print(f"  Мужское: {male_name} ({male_count})")
    print(f"  Женское: {female_name} ({female_count})")
```
